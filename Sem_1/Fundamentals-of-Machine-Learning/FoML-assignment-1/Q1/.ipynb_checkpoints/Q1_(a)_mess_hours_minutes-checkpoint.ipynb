```python
# Importing the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import warnings as w
import seaborn as sns
w.filterwarnings("ignore")
```

##### **Step 1: Reading the dataset from csv file**

```python
mess_dataset = pd.read_csv('MessFoodTimingsDataset.csv',sep=',')
print('Mess timings dataset dimensions (rows,columns):', mess_dataset.shape)
```

```python
mess_dataset.head()
```

### Exploratory Data Aalysis (EDA):

##### **(A) Feature Engineering**

Converting the 'Time' column into 'Hour' (24 hours format) & 'Minute'

```python
mess_dataset['Hour'] = pd.to_datetime(mess_dataset['Time']).dt.hour
mess_dataset['Minute'] = pd.to_datetime(mess_dataset['Time']).dt.minute
mess_dataset.drop(['Time'],axis=1,inplace=True)
mess_dataset.head()
```

##### **(E) Feature engineering:**

Feature Transformation: Applying One Hot Encoding on each categorical column

```python
categorical_column_names = mess_dataset.select_dtypes(include='object').columns
categorical_column_names
```

```python
# Applying OneHotEncoding on categorical columns
# While encoding, we are one dummy column per category to avoid dummy varible trap, i.e to avoid multicollinearity.
encoded_mess_dataset = pd.get_dummies(mess_dataset, columns = categorical_column_names)
encoded_mess_dataset.head()
```

```python
encoded_mess_dataset.shape
```

```python
corr = encoded_mess_dataset[['Time taken','Hour','Minute']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
sns.pairplot(encoded_mess_dataset[['Time taken','Hour','Minute']])
plt.show()
```

Visualizing histogram of 'Time taken' column

```python
sns.histplot(encoded_mess_dataset[['Time taken']], kde=True)
```

```python
sns.scatterplot(x='Hour', y='Time taken', data=encoded_mess_dataset)
```

##### **Step 4: Segregating the dataset into Train and Test set**

```python
X = encoded_mess_dataset.drop(columns=['Time taken'])
Y = encoded_mess_dataset['Time taken']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print('------------------ Train Set -------------------------')
print('Dimensions of X_train (rows,columns):', X_train.shape)
print('Dimensions of Y_train (rows,columns):', Y_train.shape)
print('------------------ Test Set --------------------------')
print('Dimensions of X_test (rows,columns):', X_test.shape)
print('Dimensions of Y_test (rows,columns):', Y_test.shape)
```

##### **Step 5: Fitting the Linear Regression model**

```python
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, Y_train)
```

```python
# Getting the predictions on test data
linear_reg_pred = linear_reg_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
lr_mae = mean_absolute_error(y_true=Y_test,y_pred=linear_reg_pred)
print('MAE for Linear Regression on Test set:', lr_mae)
```

```python
print('Linear regression accuracy:', round(linear_reg_model.score(X_test,Y_test) * 100,3), '%')
```

##### **Step 6: Fitting the Polynomial Regression model**

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

print(X_train_poly.shape)
print(X_train_poly)
```

```python
print('------------------ Polynomial Train Set -------------------------')
print('Dimensions of X_train (rows,columns):', X_train_poly.shape)
print('------------------ Polynomial Test Set --------------------------')
print('Dimensions of X_test (rows,columns):', X_test_poly.shape)
```

```python
lr_poly = LinearRegression()
lr_poly.fit(X_train_poly, Y_train)
```

```python
Y_pred = lr_poly.predict(X_test_poly)
poly_mae = mean_absolute_error(y_true=Y_test, y_pred=Y_pred)
```

```python
print('MAE for Polynomial Regression on Test set:', poly_mae)
print('Polynomial regression accuracy:', round(lr_poly.score(X_test_poly, Y_test) * 100,3), '%')
```
